We need to read [The List of United States Cities by Population](https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population) page on wikipedia.

```python
tables = pd.read_html('https://en.wikipedia.org/wiki/List_of_United_States_cities_by_population', header=0)
len(tables)
```

```python
tables[5]
```

```python
df_cities = tables[5].loc[:,['2019rank','City','State[c]']] # select rank, city, state columns
df_cities.rename(columns={"2019rank": "2019_rank", "State[c]": "State"}, inplace = True) # rename columns
df_cities['City'] = df_cities['City'].apply(lambda d: d.split('[')[0].strip()) #Format City name
df_cities
```
